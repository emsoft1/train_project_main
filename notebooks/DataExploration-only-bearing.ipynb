Our dataset consists of 982 individual files that are 1-second vibration signal snapshots recorded at 10 minute intervals (therefore totaling about 590000 seconds or 13 hours of observation). Each file contains 20,480 sensor data points per bearing that were obtained by reading the bearing sensors at a sampling rate of 20 kHz.

```python
# reimport libraries if only this part is run
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_path = "bearing_sensors/raw_data"
all_files = os.listdir(data_path)
sensor_col_names = [f'bearing_{i}' for i in range(1,5)]
sensor_col_names
```

```python
first_file = all_files[0]
first_file
```

```python
first_df = pd.read_csv(os.path.join(data_path, first_file), sep='\t', names=sensor_col_names)
```

```python
first_df.head(10)
```

```python
fig = plt.figure(figsize=(20,6))

plt.plot(first_df.iloc[0:1000,0])
```

```python
fig = plt.figure(figsize=(20,6))

plt.plot(first_df.iloc[:,0])
```

```python
all_dfs = []
for file_name in all_files:
    df = pd.read_csv(os.path.join(data_path, file_name), sep='\t', names=sensor_col_names)
    all_dfs.append(df)
```
